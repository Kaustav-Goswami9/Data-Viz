```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('no_encoding_data.csv')

# columns from Q4_1 to Q4_6
Q4_columns = ['Q4_1', 'Q4_2', 'Q4_3', 'Q4_4', 'Q4_5', 'Q4_6']
col_names = ['Company executives', 'Department heads', 'Immediate supervisors',
             'Internal communications', 'Large stakeholders', 'Company policies and regulations']

# create a value_counts dataframe for each column
value_counts = [data[col].value_counts() for col in Q4_columns]

# create a dataframe from the value_counts
df = pd.DataFrame(value_counts).T
df.columns = col_names
df[df.index != '-1'].to_csv('value_counts.csv')
df[df.index != '-1']
```

```python
data = pd.read_csv('no_encoding_data.csv')
q3_columns = ['Q3_1', 'Q3_2', 'Q3_3', 'Q3_4', 'Q3_5', 'Q3_6']

temp = data[data[q3_columns] != '-1']

temp = temp[q3_columns].apply(pd.Series.value_counts).T
temp.to_csv('value_counts.csv')
temp
```

```python
data = pd.read_csv('no_encoding_data.csv')
q8_columns = ['Q8_1', 'Q8_2', 'Q8_3', 'Q8_4',
              'Q8_5', 'Q8_6', 'Q8_7', 'Q8_8', 'Q8_9']

temp = data[data[q8_columns] != '-1']

temp = temp[q8_columns].apply(pd.Series.value_counts)
# change the column names
temp.columns = ['company leadership', 'union representatives', 'community council',
                'health and safety committee', 'Internal audit teams', 'grievance committee',
                'internal communications team', 'security team', "company's postal services"]
# sort by the index "A lot"
temp.sort_values(by='A lot', axis=1, ascending=False)
```
